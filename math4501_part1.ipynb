```r
# 1(a)
x=runif(10000, -1, 1)
hist(x)
```

```r
# 1(b)
U_b = runif(50000, -1, 1) #generate 50000 random samples of uniform dist from -1 to 1
#transform the vector U into a matrix of 10000 rows and 5 columns
dim(U_b) = c(10000, 5) 
M=array(0, 10000)
for(i in 1:10000)
    M[i] = max(U_b[i,])
hist(M)
```
