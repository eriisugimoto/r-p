### 1(b){-}

```r
U_b = runif(50000, -1, 1) 
dim(U_b) = c(10000, 5) 
M=array(0, 10000)
for(i in 1:10000)
    M[i] = max(U_b[i,])
    M[i]
hist(M, main = "1(b)")
```

## 1(e) {-}
I noticed that as the size of the set which we takes a max element from becomes larger, the probability of the max element being close to 1 increases. The reason for this is that the larger the size of a sample pool, the more likely it is to have all possible values so as a result, the value of the max element will become larger as the sample size increases.
