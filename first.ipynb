```r
#p.d.f of an exponential distribution with mean 1
x=seq(0, 8, 0.01)
y=dexp(x)
plot(x,y, type="l")
```

```r
#p.d.f of a normal distribution with mean 0 and standard deviation = 1
x=seq(-4, 4, 0.01)
y=dnorm(x,mean=0, sd=1)
plot(x, y, type="l")
```
